```python
"""
Exploration / Exploitation Balance:

Question: Should the agent try new actions or stick with what seems best?
Issue: Exploring can waste reward now but is needed to discover better strategies; 
       exploiting gets reward now but can miss better options.
Example: In a maze, repeatedly taking a known safe path vs trying an unknown corridor
         that might hide a shortcut.

"""
```

```python
"""
The Credit Assignment Problem:

Question: Which past action caused a reward received later?
Issue: Rewards are delayed, so it's hard to know what earlier decisions deserve credit/blame.
Example: In a game, collecting a key helps much later when opening a door.


"""
```

```python
"""
Delayed & Sparse Rewards:

Question: How can the agent learn when rewards are rare or only at the end?
Issue: With little feedback, most actions look equally useless, so learning signals are weak.
Example: Only receiving reward when solving the entire puzzle, not for intermediate progress.


"""
```

```python
"""
Partial Observability:

Question: What should the agent do when it can't see the full state?
Issue: The optimal action may depend on unobserved information and memory of past observations.
Example: In poker, you don't see opponents' cards; you must infer from betting history.


"""
```

```python
"""
Non-Stationarity:

Question: What if the environment “changes” while learning?
Issue: In RL, the data distribution changes as the policy changes; in multi-agent settings, other agents change too.
Example: You learn a tactic, then your opponent adapts, making the tactic stop working.


"""
```

```python
"""
Reward Hacking:

Question: Did we specify what we actually want?
Issue: The agent optimizes the reward literally, often finding unintended loopholes.
Example: A cleaning robot gets reward for “dirt detected” and learns to spill dirt to clean it.


"""
```
